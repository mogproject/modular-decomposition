## Benchmark Results

```python
import pandas as pd
import easy_plotly as ep
```

```python
df = pd.read_csv('bench_result.csv', header=None, names=['n', 'm', 'solver', 'mw', 'time'])
df_naive = df[df['solver'] == 'naive']
df_linear = df[df['solver'] == 'linear']
# df_naive = df[(df['solver'] == 'naive') & (df['mw'] < 100)]
# df_linear = df[(df['solver'] == 'linear') & (df['mw'] < 100)]
```

### log-scale, sorted by the number of edges, colored by modular-width

```python
fig = ep.PlotlyFigure('bench_result_scatter.yml', x_title='Number of edges', y_title='Runtime (sec)')
fig.add_scatter(x=df_naive['m'], y=df_naive['time'], marker_color=df_naive['mw'], name='py-naive', template=0)
fig.add_scatter(x=df_linear['m'], y=df_linear['time'], marker_color=df_linear['mw'], name='py-linear', template=1)
fig.to_png(show=True)
# fig.save('bench_result.png')
```
